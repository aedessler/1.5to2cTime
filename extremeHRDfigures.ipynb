```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate data for the cold‐side (blue) and hot‐side (red) curves.
x_cold = np.linspace(-10, 0, 200)
# Make the cold side steeper by dividing by a smaller number (6 instead of 10).
r_cold = 1.0 + ((-x_cold) / 6.0) ** 2

x_hot = np.linspace(0, 10, 200)
# Make the hot side shallower by using a larger denominator in the exponent (2.5 instead of 3)
# and keeping the smaller multiplier (0.1 instead of 0.2).
r_hot = 1.0 + 0.1 * (np.exp(x_hot / 2.5) - 1.0)

# 2. Set up the figure and axes.
fig, ax = plt.subplots(figsize=(8, 4))

# Plot the two curves.
ax.plot(x_cold, r_cold, color='blue', linewidth=2)
ax.plot(x_hot, r_hot, color='red', linewidth=2)

# 3. Draw a black dot at the minimum point (x=0, risk=1).
ax.scatter([0], [1.0], color='black', s=50, zorder=5)

# 4. Hide the top and right spines.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Draw the horizontal dotted line at y = 1.
ax.axhline(1.0, color='black', linestyle=':', linewidth=1)

# 5. Shade the region between r_hot and the dotted line for x > 0.
ax.fill_between(
    x_hot,
    r_hot,
    1.0,
    where=(r_hot >= 4.0),
    facecolor='red',
    alpha=0.2,
    interpolate=True
)

# 6. Remove tick marks.
ax.set_xticks([])
ax.set_yticks([])

# 7. Set axis limits so everything is visible.
ax.set_xlim(-10.5, 10.5)
y_max = max([np.max(r_cold), np.max(r_hot)]) + 0.5
ax.set_ylim(-1.0, y_max)

# 8. Label the y‐axis.
ax.set_ylabel('number of deaths in a day', fontsize=12, labelpad=15)

# 9. Place “Cold” and “Hot” along the bottom.
ax.text(-10.0, -1.2, 'Cold', color='blue',
        horizontalalignment='left', verticalalignment='top', fontsize=12)
ax.text(10.0, -1.2, 'Hot', color='red',
        horizontalalignment='right', verticalalignment='top', fontsize=12)

# 10. Label the x‐axis.
ax.set_xlabel('temperature', fontsize=12, labelpad=15)

# 11. Save and show.
plt.tight_layout()
plt.savefig('risk_shaded_xhm.png', dpi=300)
plt.show()
```
